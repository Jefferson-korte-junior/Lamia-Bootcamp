```python
 # Configura o ambiente do Jupyter Notebook para exibir gráficos embutidos
%matplotlib inline 

import pandas as pd  # Importa a biblioteca pandas, usada para manipulação e análise de dados

# Lê um arquivo CSV a partir de uma URL e armazena os dados em um DataFrame 'df'
df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

# Conta o número de ocorrências de cada valor na coluna '# Gears' do DataFrame 'df' e armazena em 'gear_counts'
gear_counts = df['# Gears'].value_counts()

# Cria um gráfico de barras (bar plot) a partir dos dados em 'gear_counts'
gear_counts.plot(kind='bar')
```

```python
 # Importa a biblioteca Seaborn, que é usada para criar visualizações estatísticas de dados
import seaborn as sns 

# Configura os temas estéticos padrão para todos os gráficos subsequentes criados com Seaborn e Matplotlib
sns.set()  
```

```python
# Cria um gráfico de barras a partir da contagem de valores em 'gear_counts'
gear_counts.plot(kind='bar')  
```

```python
# Exibe as primeiras 5 linhas do DataFrame 'df'
df.head()  
```

```python
# Cria um gráfico de distribuição para a coluna 'CombMPG' do DataFrame 'df'
sns.distplot(df['CombMPG'])  
```

```python
# Cria um novo DataFrame 'df2' contendo apenas as colunas 'Cylinders', 'CityMPG', 'HwyMPG' e 'CombMPG' do DataFrame original 'df'
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]  

# Exibe as primeiras 5 linhas do novo DataFrame 'df2'
df2.head()  
```

```python
# Cria uma matriz de gráficos de dispersão (scatter plots) para cada par de colunas em 'df2', com cada gráfico de 2,5 unidades de altura
sns.pairplot(df2, height=2.5);  
```

```python
 # Cria um gráfico de dispersão (scatter plot) comparando 'Eng Displ' (deslocamento do motor) e 'CombMPG' (consumo combinado de combustível) usando os dados do DataFrame 'df'
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df) 
```

```python
# Cria um gráfico de dispersão (scatter plot) com uma linha de regressão linear para 'Eng Displ' e 'CombMPG' usando os dados do DataFrame 'df'
sns.lmplot(x="Eng Displ", y="CombMPG", data=df)  
```

```python
# Configura o tamanho padrão das figuras para todos os gráficos subsequentes, definindo a largura como 15 e a altura como 5 polegadas
sns.set(rc={'figure.figsize':(15,5)})  


 # Cria um gráfico de caixa (box plot) para 'CombMPG' em relação a 'Mfr Name' usando os dados do DataFrame 'df'
ax = sns.boxplot(x='Mfr Name', y='CombMPG', data=df) 


 # Ajusta os rótulos do eixo x para que sejam rotacionados 45 graus, melhorando a legibilidade
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) 
```

```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')  
# Cria uma tabela dinâmica (pivot table) a partir do DataFrame 'df', onde:
# - 'index' define a coluna 'Cylinders' como o índice da tabela
# - 'columns' define a coluna 'Eng Displ' como as colunas da tabela
# - 'values' especifica a coluna 'CombMPG' cujos valores serão agregados
# - 'aggfunc' determina a função de agregação, que neste caso é a média ('mean')

sns.heatmap(df2)  
# Cria um mapa de calor a partir do DataFrame 'df2' criado anteriormente

# O mapa de calor visualiza a média do 'CombMPG' para diferentes combinações de 'Cylinders' e 'Eng Displ'
```
