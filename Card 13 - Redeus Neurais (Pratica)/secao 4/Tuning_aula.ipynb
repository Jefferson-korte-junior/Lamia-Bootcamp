```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
from sklearn.model_selection import GridSearchCV
#O GridSearchCV automatiza a busca pela melhor configuração do seu modelo, avaliando cada opção com validação cruzada para garantir que o desempenho seja consistente — e não só sorte de uma única divisão dos dados.
```

```python
X = pd.read_csv('entradas_breast_aula.csv')
Y = pd.read_csv('saidas_breast_aula.csv')
```

```python
def criar_rede(optimizer, loss, kernel_initializer, activation, neurons):
    tf.compat.v1.reset_default_graph()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(input_shape=(30,)),
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=1, activation="sigmoid")
    ])
    rede_neural.compile(optimizer=optimizer, loss=loss, metrics=["binary_accuracy"])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede)
```

```python
parametros = {
    'batch_size': [10,30],
    'epochs': [50],
    'model__optimizer': ['adam'],
    'model__loss': ['binary_crossentropy'],
    'model__kernel_initializer': ['random_uniform', 'normal'],
    'model__activation': ['relu'],
    'model__neurons': [16]
}
```

```python
grid_search = GridSearchCV(estimator = rede_neural, param_grid = parametros, scoring = 'accuracy', cv = 5)
```

```python
grid_search
```

```python
grid_search = grid_search.fit(X, Y)
```

```python
print(grid_search)
```

```python
melhores_parametros = grid_search.best_params_
print(melhores_parametros)
```

```python
melhor_precisao = grid_search.best_score_
print(melhor_precisao)
```
