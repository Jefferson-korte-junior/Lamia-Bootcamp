```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as K
```

```python
# Importa a função fetch_ucirepo da biblioteca ucimlrepo,
# que serve para buscar datasets diretamente do UCI Machine Learning Repository.
from ucimlrepo import fetch_ucirepo

# Busca o dataset com id 17, que corresponde ao "Breast Cancer Wisconsin (Diagnostic)".
# Esse conjunto de dados é usado para prever se tumores são benignos ou malignos.
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

# Seleciona os atributos (features) do dataset, ou seja,
# variáveis como raio médio, textura, simetria etc., que serão usadas como entrada para um modelo.
X = breast_cancer_wisconsin_diagnostic.data.features

# Seleciona os rótulos (targets), que indicam o diagnóstico real:
# benigno ou maligno. Essa será a saída que o modelo tentará prever.
Y = breast_cancer_wisconsin_diagnostic.data.targets
```

```python
from scikeras.wrappers import KerasClassifier

#define a arquitetura de uma rede neural para classificação binária.

def criar_rede():
    tf.compat.v1.reset_default_graph()
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(shape = (30,)),
        tf.keras.layers.Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'),
        tf.keras.layers.Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'),
        tf.keras.layers.Dense(units = 1, activation = 'sigmoid')])
    otimizador = tf.keras.optimizers.Adam(learning_rate = 0.001, clipvalue = 0.5)
    rede_neural.compile(optimizer = otimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
    return rede_neural
```

```python
rede_neural =  KerasClassifier(model = criar_rede, epochs= 100, batch_size= 10)
```

```python
resultado = cross_val_score(estimator = rede_neural, X = X, y = Y, cv = 10, scoring = "accuracy")
```
