```python
base['brand'].value_counts()
```

```python
onehotencoder = ColumnTransformer(
    transformers=[('OneHot', OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])],  # Converte as colunas categóricas (como textos) em números binários
    remainder='passthrough'  # Mantém as outras colunas do jeito que estão
)
```

```python
X = onehotencoder.fit_transform(X).toarray()
X
```

```python
base.shape
```

```python
X.shape
```

```python
regressor = Sequential([
    tf.keras.layers.InputLayer(shape = (316,)),
    tf.keras.layers.Dense(units = 158, activation = 'relu'),
    tf.keras.layers.Dense(units = 158, activation = 'relu'),
    tf.keras.layers.Dense(units = 1, activation = 'linear'),
])
```

```python
regressor.summary()
```

```python
regressor.compile(loss = 'mean_absolute_error', optimizer= 'adam', metrics= ['mean_absolute_error'])
```

```python
regressor.fit(X, Y, batch_size = 300, epochs = 100)
```

```python
previsoes = regressor.predict(X)
```

```python
previsoes, Y
```
