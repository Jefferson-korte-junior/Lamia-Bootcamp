Funçao step, é uma funçao linear

```python
def step_function(soma):
  if soma>= 1:
      return 1
  return 0
```

```python
step_function(-10)
```

Funçao sigmoide vai retorna um numero entre o e 1

```python
def sigmoid_function(soma):
  return 1/(1+np.exp(-soma))
```

```python
sigmoid_function(10) # Quando passamos o valor 10, ele retorna a probabilidade desse valor ser positov
```

```python
sigmoid_function(0.1)
```

Função tangente hiperbólica, podem ser retornados valores entre -1 e 1, mais utilizado em visão computacional

```python
def tahn_function(soma):
  return (np.exp(soma)-np.exp(-soma))/(np.exp(soma)+np.exp(-soma))
```

tahn_function tem a capacidade de retornar valores negativos para entradas negativas, o que a sigmoid_function não tem. Isso pode ser útil em modelos onde é importante ter saídas centralizadas em torno de zero.

```python
tahn_function(-10) # retorna um número entre -1 e 1. Assim como a sigmoide mas seu intervalo de saída de -1 a 1.
```

Função Relu (Rectified Linear Units), valores>=0 (sem valor máximo). Mais utilizados em Deep Learning

```python
def relu_function(soma):
  if soma >= 0:
    return soma
  return 0
```

```python
relu_function(10)
```

Soft max

```python
def softmax_function(x):
    exp = np.exp(x)
    return exp / exp.sum()
```

```python
softmax_function([7.0, 2.0, 1.1, 7.2, 7.1])
```
