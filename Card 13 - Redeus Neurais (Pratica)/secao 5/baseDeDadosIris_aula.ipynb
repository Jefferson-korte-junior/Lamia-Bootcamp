```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
from tensorflow.keras.models import Sequential  # Importa o modelo sequencial para criar redes neurais camada por camada
from tensorflow.keras.layers import Dense       # Importa a camada densa (totalmente conectada)
from tensorflow.keras import utils as np_utils  # Importa utilitários do Keras, como o to_categorical
from sklearn.preprocessing import LabelEncoder  # Importa o codificador para transformar rótulos em valores numéricos
from sklearn.model_selection import train_test_split  # Importa a função para dividir os dados em treino e teste
```

```python
base = pd.read_csv('iris_aula.csv')
base.head()
```

```python
X = base.iloc[:, 0:4].values
```

```python
Y = base.iloc[:, 4].values
```

```python
label_encoder = LabelEncoder()
Y = label_encoder.fit_transform(Y)
Y
```

```python
Y = np_utils.to_categorical(Y)
Y
```

```python
X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size=0.25)
```

```python
X_treinamento.shape, X_teste.shape
```

```python
Y_treinamento.shape, Y_teste.shape
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units=3, activation = "softmax"),
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede_neural.fit(X_treinamento, Y_treinamento, batch_size = 10, epochs = 1000)
```

```python
rede_neural.evaluate(X_teste, Y_teste)
```

```python
previsoes = rede_neural.predict(X_teste)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```
