```python
import numpy as np
from matplotlib.pyplot import scatter
from matplotlib import pyplot as plt
```

```python
# Gera um array com 1000 valores de tempos de carregamento de página (em segundos)
# Os valores seguem uma distribuição normal com média 3.0 e desvio padrão 1.0
PageSpeeds = np.random.normal(3.0, 1.0, 1000)

# Gera o valor de compra para cada usuário, com base na velocidade da página
# Quanto mais lenta a página (PageSpeeds maior), menor será o valor da compra
# Também adiciona um pequeno ruído aleatório com média 0 e desvio 0.1 para dar realismo
purchaseAmount = 100 - (PageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

# Plota um gráfico de dispersão (scatter plot)
# Eixo X: tempo de carregamento da página
# Eixo Y: valor da compra
scatter(PageSpeeds, purchaseAmount)

plt.xlabel('Tempo de carregamento da página (segundos)')
plt.ylabel('Valor da compra')
plt.title('Relação entre tempo de carregamento e valor da compra (baixa variância)')
```

```python
from scipy import stats  # Importa o módulo 'stats' da biblioteca SciPy, que contém funções estatísticas

# Calcula a regressão linear entre as variáveis 'pageSpeeds' (tempo de carregamento) e 'purchaseAmount' (valor da compra)

# slope      → Inclinação da reta de regressão (o quanto o valor da compra varia a cada segundo a mais de carregamento)
# intercept  → Intercepto da reta (valor da compra quando o tempo de carregamento é zero)
# r_value    → Coeficiente de correlação (indica a força e direção da relação entre as variáveis)
# p_value    → Valor-p do teste de hipótese (indica se a relação é estatisticamente significativa)
# std_err    → Erro padrão da inclinação (mede a incerteza da inclinação estimada)
slope, intercept, r_value, p_value, std_err = stats.linregress(PageSpeeds, purchaseAmount)
```

```python
r_value ** 2 #é um resultado muito bom para a regressao
```

```python
# Importa a biblioteca matplotlib.pyplot com o apelido 'plt', usada para criar gráficos
import matplotlib.pyplot as plt

# Define uma função chamada 'predict' que calcula o valor predito (y) para um dado x, com base na equação da reta de regressão
# Equação da reta: y = slope * x + intercept
def predict(x):
    return slope * x + intercept

# Gera os valores da linha de regressão (linha de melhor ajuste) para todos os pontos de 'pageSpeeds'
# Isso cria um array com os valores de y previstos (fitLine), baseado nos valores de x (pageSpeeds)
fitLine = predict(PageSpeeds)

# Cria um gráfico de dispersão (scatter plot) mostrando os dados reais:
# Eixo X: tempo de carregamento da página (pageSpeeds)
# Eixo Y: valor da compra (purchaseAmount)
plt.scatter(PageSpeeds, purchaseAmount)

# Plota a linha de regressão (linha de melhor ajuste) em vermelho (c='r')
# Essa linha representa o modelo linear ajustado aos dados
plt.plot(PageSpeeds, fitLine, c='r')

plt.xlabel('Tempo de carregamento da página (segundos)')
plt.ylabel('Valor da compra')
plt.title('Relação entre tempo de carregamento e valor da compra (baixa variância)')

# Exibe o gráfico na tela
plt.show()
```

# Tente aumentar a variação aleatória nos dados de teste e veja qual efeito isso tem no valor do erro R-quadrado

```python
# Gera um array com 1000 valores de tempos de carregamento de página (em segundos)
# Os valores seguem uma distribuição normal com média 3.0 e desvio padrão menor (0.2) para reduzir variância
PageSpeeds = np.random.normal(3.0, 0.1, 1000)

# Gera o valor de compra para cada usuário, com base na velocidade da página
# Ruído menor (0.02) para menor variância na compra também
purchaseAmount = 100 - (PageSpeeds + np.random.normal(0, 0.08, 1000)) * 3

# Plota um gráfico de dispersão (scatter plot)
# Eixo X: tempo de carregamento da página
# Eixo Y: valor da compra
plt.scatter(PageSpeeds, purchaseAmount)
plt.xlabel('Tempo de carregamento da página (segundos)')
plt.ylabel('Valor da compra')
plt.title('Relação entre tempo de carregamento e valor da compra (baixa variância)')
plt.show()
```

```python
from scipy import stats  # Importa o módulo 'stats' da biblioteca SciPy, que contém funções estatísticas

# Calcula a regressão linear entre as variáveis 'pageSpeeds' (tempo de carregamento) e 'purchaseAmount' (valor da compra)

# slope      → Inclinação da reta de regressão (o quanto o valor da compra varia a cada segundo a mais de carregamento)
# intercept  → Intercepto da reta (valor da compra quando o tempo de carregamento é zero)
# r_value    → Coeficiente de correlação (indica a força e direção da relação entre as variáveis)
# p_value    → Valor-p do teste de hipótese (indica se a relação é estatisticamente significativa)
# std_err    → Erro padrão da inclinação (mede a incerteza da inclinação estimada)
slope, intercept, r_value, p_value, std_err = stats.linregress(PageSpeeds, purchaseAmount)
```

```python
r_value ** 2 #é um resultado muito ruim para fins de regressão. Esse valor indica que menos de 0,1% da variação no valor da compra é explicada pela velocidade da página.
```

```python
# Define uma função chamada 'predict' que calcula o valor predito (y) para um dado x, com base na equação da reta de regressão
# Equação da reta: y = slope * x + intercept
def predict(x):
    return slope * x + intercept

# Gera os valores da linha de regressão (linha de melhor ajuste) para todos os pontos de 'pageSpeeds'
# Isso cria um array com os valores de y previstos (fitLine), baseado nos valores de x (pageSpeeds)
fitLine = predict(PageSpeeds)

# Cria um gráfico de dispersão (scatter plot) mostrando os dados reais:
# Eixo X: tempo de carregamento da página (pageSpeeds)
# Eixo Y: valor da compra (purchaseAmount)
plt.scatter(PageSpeeds, purchaseAmount)

# Plota a linha de regressão (linha de melhor ajuste) em vermelho (c='r')
# Essa linha representa o modelo linear ajustado aos dados
plt.plot(PageSpeeds, fitLine, c='r')

plt.xlabel('Tempo de carregamento da página (segundos)')
plt.ylabel('Valor da compra')
plt.title('Relação entre tempo de carregamento e valor da compra (baixa variância)')

# Exibe o gráfico na tela
plt.show()
```

Eu peguei e aumentei a variância aumentando o desvio padrão do ruído nos valores de compra (de 0.02 para 0.08), o que deixou os pontos mais espalhados no eixo Y. Além disso, diminui o desvio padrão de PageSpeeds (de 0.2 para 0.1), mantendo os valores de carregamento mais concentrados. O principal impacto foi no aumento da dispersão vertical (valores de compra) no gráfico.
