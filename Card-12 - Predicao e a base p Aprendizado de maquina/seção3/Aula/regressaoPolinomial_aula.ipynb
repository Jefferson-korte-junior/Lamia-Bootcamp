```python
import numpy as np
import matplotlib.pyplot as plt

# Define a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(2)

# Gera 1000 valores de velocidades de carregamento de página com média 3.0 e desvio padrão 1.0
pageSpeeds = np.random.normal(3.0, 1.0, 1000)

# Gera 1000 valores de gasto com média 50.0 e desvio padrão 10.0, e divide pela velocidade da página
# Isso cria uma correlação inversa entre a velocidade da página e o valor da compra
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

# Cria um gráfico de dispersão dos dados gerados
plt.scatter(pageSpeeds, purchaseAmount)

# Exibe o gráfico
plt.show()
```

```python
# Converte a lista de velocidades de página para um array NumPy (caso ainda não seja)
x = np.array(pageSpeeds)

# Converte a lista de valores de compra para um array NumPy
y = np.array(purchaseAmount)

# Ajusta um polinômio de grau 4 (quarto grau) aos dados de x e y
# np.polyfit encontra os coeficientes do polinômio que melhor se ajusta aos dados
# np.poly1d cria um objeto polinomial com esses coeficientes, facilitando o uso posterior
p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
# Cria 100 pontos igualmente espaçados entre 0 e 7
# Esses pontos servirão como eixo x para desenhar a curva do polinômio
xp = np.linspace(0, 7, 100)

# Plota os dados originais em um gráfico de dispersão
plt.scatter(x, y)

# Plota a curva do polinômio de grau 4 (usando os valores de xp no eixo x)
# p4(xp) calcula os valores y do polinômio correspondente a cada ponto xp
# c='r' define a cor da linha como vermelha (red)
plt.plot(xp, p4(xp), c='r')

# Exibe o gráfico final com os dados e a curva ajustada
plt.show()
```

```python
# Importa a função r2_score da biblioteca scikit-learn
# Ela é usada para calcular o coeficiente de determinação (R²), que mede o quão bem o modelo se ajusta aos dados
from sklearn.metrics import r2_score

# Calcula o valor de R² entre os valores reais (y) e os valores previstos pelo polinômio p4 aplicado a x
# Quanto mais próximo de 1, melhor o ajuste do modelo aos dados
r2 = r2_score(y, p4(x))

# Exibe o valor de R² no console
print(r2)
```
